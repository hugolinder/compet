```python
import kex_headers
import numpy as np
import matplotlib.pyplot as plt
```

```python
seg_sizes = kex_headers.SEGMENT_TABLE
plt.plot(seg_sizes)
plt.ylabel('segment size')
plt.xlabel('segment bin');
```

```python
max_mi = np.cumsum(seg_sizes)
min_mi = max_mi - seg_sizes
plt.plot(max_mi, label = 'max_mi')
plt.plot(min_mi, label = 'min_mi')
plt.legend()
plt.xlabel('segment bin')
plt.ylabel('mi bin');
```

```python
mi = np.arange(621)
plt.plot(np.searchsorted(max_mi-1, mi), label='search max')
plt.xlabel('michlogram bin')
plt.ylabel('segment bin');
```
