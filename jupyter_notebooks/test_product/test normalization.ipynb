```python
import position
import numpy as np
import matplotlib.pyplot as plt
```

```python
weights = np.arange(10)
limo = []
for w in weights:
    limo = limo + [w]*w
plt.plot(limo)
plt.show()
norm_masses = 1/weights[limo]
plt.plot(norm_masses)
```

```python
norm_com = position.listmode_center(np.array([limo]), masses=norm_masses)
com = position.listmode_center(np.array([limo]))
print("com", com)
print("normalized com", norm_com)
```
