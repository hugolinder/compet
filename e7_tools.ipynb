```python
import os
from subprocess import Popen, PIPE

# e7 folder on author's laptop
e7_FOLDER = r"C:/Users/petct/Desktop/Siemense7tools/Siemense7tools/C-Siemens-PET-VG60/bin.win64-VG60/"
```

```python
def run_e7_tool(tool_str, parameters=[], e7_folder=None, verbose = True):
    """ runs the e7 tool with the given parameters. If tool_str does not end with .exe, that is concatenated. 
    No tool parameters yields the documentation. Prints program standard output if verbose."""
    
    if not tool_str.endswith(".exe"):
        tool_str = tool_str + ".exe"
    if e7_folder is None:
        program_path = e7_FOLDER + r"/" + tool_str 
    else:
        program_path = e7_folder + r"/" + tool_str 
    
    process = Popen([program_path] + parameters, stdout=PIPE, stderr=PIPE)
    if verbose:
        stdout, stderr = process.communicate()

        stdout_str = str(stdout)
        print(stdout_str.replace(r"\r\n", "\n"))
    return
```
